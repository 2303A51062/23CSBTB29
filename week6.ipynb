Dataset: E-commerce Sales Data

Load dataset and display first three rows

```python
import pandas as pd

order_id_col = 'Order_ID'

data = {
    order_id_col: [1001, 1002, 1003, 1004],
    "Product": ["Desktop", "Smartphone", "iPad", "Gaming Laptop"],
     "Price ($)":[5000,None,600,8200],
    "Quantity": [1, 2, None, 1],
    "Discount (%)": [10, 5, 0, None]
}

df = pd.DataFrame(data)
print(df.head(3))
```

Identify missing values and count them

```python
print(df.isnull().sum()) # Counts missing values
```

Fill missing Price with mean and missing Quantity with 1

```python
df['Price ($)'].fillna(df['Price ($)'].mean(), inplace=True)
df['Quantity'].fillna(1, inplace=True)
```

Calculate total revenue

```python
df['Revenue'] = (df['Price ($)'] * df['Quantity']) - df['Discount (%)'].fillna(0)
print(df[['Order_ID', 'Revenue']])
```
